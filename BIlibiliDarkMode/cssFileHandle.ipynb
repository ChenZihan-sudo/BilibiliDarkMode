```python
fileName = "n_patch.css"  # 本地目录文件名 Local Dir File Name
varCssFileName = "root.css"

import os
import re
import math


def getFileName(fileName):
    path = os.getcwd() + "\\" + fileName
    path = re.sub(r'\\', r'\\\\', path)
    return path


def clearDataArrayAddCssImportant(cssData):
    while 1:
        searchReg = re.search(r'\[data-v-[qwwqeqwe0-9a-z]*\]', cssData)

        if searchReg != None:
            beginPos = searchReg.span()[0]
            endPos = searchReg.span()[1]

            i = endPos
            while i < len(cssData):
                i += 1
                if cssData[i] == ';':
                    str1 = cssData[:i]
                    str2 = "!important;" + cssData[i + 1:]
                    cssData = str1 + str2
                    i = i + len("!important;")
                elif cssData[i] == '}':
                    str1 = cssData[0:beginPos]
                    str2 = cssData[endPos:]
                    cssData = str1 + str2
                    break
        else:
            break
    return cssData


def getDictOfVarCss(varCssFileName):
    varCssFileNamePath = getFileName(varCssFileName)
    varStrData = ""

    varData = dict()

    with open(varCssFileNamePath, 'r') as f:
        varStrData = f.read()

    eachEndPosi = 0

    while 1:
        searchReg = re.search(r'--[-a-z0-9A-Z]*:', varStrData[eachEndPosi:])

        if searchReg == None:
            break

        titleMatchData = searchReg.group()[:-1]

        titleMatchBeginPos = searchReg.span()[0] + eachEndPosi
        titleMatchEndPos = searchReg.span()[1] + eachEndPosi

        colorMatchBeginPos = -math.inf
        colorMatchEndPos = -math.inf

        for i in range(titleMatchEndPos, len(varStrData)):
            if varStrData[i] == '#':
                colorMatchBeginPos = i
                break

        hexDict = "0123456789abcdefABCDEF"
        for i in range(colorMatchBeginPos + 1, len(varStrData)):
            isEndOfColor = True
            for j in range(len(hexDict)):
                if varStrData[i] == hexDict[j]:
                    isEndOfColor = False
                    break
            if isEndOfColor == True:
                colorMatchEndPos = i
                break

        varData[titleMatchData] = varStrData[
            colorMatchBeginPos:colorMatchEndPos]
        eachEndPosi = colorMatchEndPos
    return varData


def tranColorToVarColor(varData, targetData):
    for key in varData:
        while 1:
            # targetData = re.sub(varData[key], "var(" + key + ")", targetData)

            searchReg = "[" + varData[key] + "]{" + str(len(
                varData[key])) + "}[ !important;]*$"

            searchRegData = re.search(searchReg, targetData, re.M)

            if searchRegData != None:
                # searchRegDataStr__ = searchRegData
                # print(searchRegDataStr__)
                searchRegDataStr = searchRegData.group()
                checkStr = searchRegDataStr.find(
                    ';') != -1 or searchRegDataStr.find('!') != -1

                if True:
                    searchRegDataBeginPos = searchRegData.span()[0]
                    searchRegDataEndPos = searchRegData.span()[1]
                    Str1 = "var(" + key + ")"
                    Str2 = searchRegDataStr[len(varData[key]):]

                    readyStr1 = targetData[:searchRegDataBeginPos]
                    readyStr2 = Str1 + Str2
                    readyStr3 = targetData[searchRegDataEndPos:]

                    targetData = readyStr1 + readyStr2 + readyStr3
            else:
                break
    return targetData


filePath = getFileName(fileName)
cssData = ""
with open(filePath, 'r+') as f:
    cssData = f.read()

cssData = clearDataArrayAddCssImportant(cssData)

varData = getDictOfVarCss(varCssFileName)
finalData = tranColorToVarColor(varData, cssData)

fileWriteTargetPath = getFileName("Processed_" + fileName)
with open(fileWriteTargetPath, 'w') as f:
    f.write(finalData)
```
